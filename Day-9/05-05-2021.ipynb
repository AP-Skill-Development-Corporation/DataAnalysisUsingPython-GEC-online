### Area Plot or Stack plot
- plt.stackplot()

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]
age = [40,70,82,16,33]
weight = [65,30,72,44,55]
plt.stackplot(days,age,weight)
plt.show()
```

```python
plt.stackplot(days,age,weight, colors = ['g','m'], labels = ['Age','Weight'])
plt.legend()
```

### Save the Graph in folder

```python
plt.stackplot(days,age,weight, colors = ['g','m'], labels = ['Age','Weight'])
plt.legend()
plt.savefig('areaplot.png')
```

### 1. Color Palette
- It is an interface to generate different shades of colors in Seaborn
- sns.color_palette()

```python
sns.color_palette()
```

```python
help(sns.color_palette)
```

```python
sns.palplot(sns.color_palette())
```

### Available seaborn palette names:
- deep, muted, bright, pastel, dark, colorblind

```python
sns.palplot(sns.color_palette('deep'))
```

```python
sns.palplot(sns.color_palette('dark'))
```

```python
sns.palplot(sns.color_palette('colorblind'))
```

```python
sns.palplot(sns.color_palette("RdBu"))
```

```python
sns.palplot(sns.color_palette("RdBu", n_colors=10))
```

```python
sns.palplot(sns.color_palette("purple"))
```
