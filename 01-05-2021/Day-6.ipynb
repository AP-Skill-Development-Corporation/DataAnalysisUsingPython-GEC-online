### Replace Null values using fillna()
- If data set is numerical, Relace the Null values by using zero or mean().
- If data set is categorical, Relace the Null values by using most occured object name.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Global%20Dataset/Market_Fact.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Product_Base_Margin'].mean()
```

```python
df['Product_Base_Margin'] = df['Product_Base_Margin'].fillna(df['Product_Base_Margin'].mean())
```

```python
df.isnull().sum()
```

### Outliers
- Outliers are different from other data points, which will effects the training process of machine learning algorithms and resulting in a loss of accuracy
- For handling outliers, apply scaling techniques.

### Scaling Techniques 
- Standard scaler
- Min Max Scaler
- Robust scaler
- etc,..

```python
import pandas as pd
df2 = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
df2.head()
```

### 1. Standard Scaler
- If data is normally distributed, apply standard scaler
- It scales all the data points in between 0 to std value
    - Z = [Xi - mean(X) ] / std(X)

```python
plt.hist(df2)
plt.show()
```

```python
## Import required packages
from sklearn.preprocessing import StandardScaler
```

```python
## Assign model to object
ss = StandardScaler()
```

```python
df2.describe()
```

```python
(230.1 - 147.042500) / 85.854236
```

```python
## By fitting model, it will apply formula to dataset
ss2 = ss.fit_transform(df2)
ss2
```

```python
## Convert array values into dataframe
ss2 = pd.DataFrame(ss2, columns = df2.columns)
ss2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))

sns.kdeplot(df2['TV'], ax = ax1)
sns.kdeplot(df2['radio'], ax = ax1)
sns.kdeplot(df2['newspaper'], ax = ax1)
sns.kdeplot(df2['sales'], ax = ax1)

sns.kdeplot(ss2['TV'], ax = ax2)
sns.kdeplot(ss2['radio'], ax = ax2)
sns.kdeplot(ss2['newspaper'], ax = ax2)
sns.kdeplot(ss2['sales'], ax = ax2)
```

### 2. Robust Scaler
- If your data set contains huge amount of outliers, Robust scaler is the best scaling technique
- Scaling between Inter Ouartile Range(25% to 75%)
    - [ Xi - Q1(X) ] / [ Q3(X) - Q1(X) ]
        - Q1 - 25%(Lower Quartile)
        - Q2 - 50%(Mediun)
        - Q3 - 75%(Upper quartile)
